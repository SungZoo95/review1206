```python
# 계단 함수 구현해보기
def step_function(x):
    if x > 0:
        return 1 # x가 0보다 크면 1
    else: 
        return 0 # x가 0보다 작으면 0 기본형
```

```python
import numpy as np
x_1 = np.array([-1.0,1.0,-2.0,0.5])
y_1 = x_1 > 0 
print(y_1)
print(type(y_1) )
print(y_1.astype(int))
type(y_1.astype(int))
```

```python
import numpy as np
def step_function_1(x):
    y = x > 0
    return y.astype(int)
```

```python
x_2 = np.array([-1.0, 1.0, -2.0])
y_2 = step_function_1(x_2)
```

```python
import matplotlib.pyplot as plt
x_3 = np.arange(-5.0, 5.0, 0.1)
y_3 = step_function_1(x_3)
plt.plot(x_3,y_3)
plt.ylim(-0.1,1.1)
plt.show()
```

```python
# 시그모이드 함수 구현
def sigmoid(x):
    
    return 1/(1+np.exp(-x))
```

```python
x_4 = x_3
y_4 = sigmoid(x_4)
plt.plot(x_4, y_4)
plt.ylim(-0.1,1.1)
plt.show()
```
