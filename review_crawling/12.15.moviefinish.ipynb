```python
df = pd.DataFrame(columns = ["번호", "제목", "감상평", "점수", "점수2"])

for j in range(1, 10):
    naver_mov = f"https://movie.naver.com/movie/point/af/list.naver?&page={j+1}"
    naver_req = requests.get(naver_mov)
    table_pd = pd.read_html(naver_req.text)
    
    for i in range(1, 10):
        
        movie_num = table_pd[0]['번호'][i]
        movie_title = table_pd[0]['감상평'][i].split(" 별점 - 총 10점 중")
        movie_write = table_pd[0]['감상평'][i].split(" 별점 - 총 10점 중")[1][:-3].split("  ")
        star = table_pd[0]['감상평'][i].split(" 별점 - 총 10점 중")[1][:-3].split("  ")[0]
        style = star
        if int(star) >= 7:
            style = 1
        else : 
            style = 0 
        if len(movie_write) == 2:
            df =  df.append({"번호": int(movie_num), "제목": movie_title[0], "감상평": movie_write[1], "점수": star, "점수2":style}
                ,  ignore_index=True)
       
        
    
```
